### ベクトルと1次元配列

- ベクトルのような複数の数値が集まったものを1次元配列という。

$$
\mbox{列ベクトル}:\vec{x}=\left(
\begin{array}{c}
1 \\
2 \\
3 \\
\end{array}
\right)
$$

このような縦方向に数値が並んだベクトルを列ベクトルという。

一方、横方向に数値が並んだベクトルを行ベクトルという。

$$
\mbox{行ベクトル}:\vec{y}=(1,2,3)
$$

行ベクトルを列ベクトルに、また、列ベクトルを行ベクトルに変換するには転置という操作を行います。ベクトルの右上に$T$と書きます。

$$
\mbox{転置}:\vec{x}=\left(
\begin{array}{c}
1 \\
2 \\
3 \\
\end{array}
\right)\to \vec{x}^T=(1,2,3)=\vec{y}
$$

$$
\mbox{転置}:
\vec{y}=(1,2,3) \to \vec{y}^T\left(
\begin{array}{c}
1 \\
2 \\
3 \\
\end{array}
\right)=\vec{x}
$$






```python
# 成分が(1,2,3)のベクトルxを作ります。
x=np.array([1,2,3])
x
```

## 決定境界

パーセプトロンは$w_{0}+w_{1}x+w_{2}y$を計算して正なら1,0以下なら0と判定します。

$$
\phi(x,y)=\left\{
\begin{array}{cc}
1 & w_{0}+w_{1}x+w_{2}y>0 \\
0 & w_{0}+w_{1}x+w_{2}y \leqq 0 \\
\end{array}
\right.
$$

1と0を判定する各領域の境界を決定境界と呼びます。決定境界は次の
関係式を満たします。

$$
w_{0}+w_{1}x+w_{2}y=0
$$

$y$を解くと

$$
y=-\frac{w_1}{w_2}x-\frac{w_0}{w_2}
$$

となります。つまり、決定境界は直線です。

### ANDの決定境界

$w_0=-3,w_1=2,w_2=2$とするとパーセプトロンはANDの計算が出来ます。
このとき、決定境界は

$$
y=-\frac{w_1}{w_2}x-\frac{w_0}{w_2}=-x-\frac{3}{2}
$$

となります。

ANDのデータと決定境界をプロットしてみます。


```python
# プロットする図の大きさをfigsize=(12,8)で指定します。
plt.figure(figsize=(12,8)) 
# OUT=1に対応するXの行のデータを2次元面にプロットします。
# xにX[,0],yにX[,1]をインプット、maker="s"としてマーカーを"四角"、サイズをs=200、色をcolor="red"と指定します。
plt.scatter(x=AND[OUT_AND==1,0],y=AND[OUT_AND==1,1],label='OUT=1',marker="s",s=200,color="red")
# OUT=0に対応するXの行のデータを2次元面にプロットします。
# xにX[,0],yにX[,1]をインプット、maker="o"としてマーカーを"丸"、サイズをs=200、色をcolor="blue"と指定します。
plt.scatter(x=AND[OUT_AND==0,0],y=AND[OUT_AND==0,1],label='OUT=0',marker="o",s=200,color="blue")
# 決定境界をプロットするために、x軸の2点を選びます。
xs=np.array([-0.5,1.5])
# 決定境界のy座標を計算します。
ys=xs*(-1)+1.5
# 決定境界の座標を直線で結びます。
plt.plot(xs,ys,linewidth=5,color="green")
# 表示するx軸の範囲を指定します。
plt.xlim([-0.5,1.5])
# 表示するy軸の範囲を指定します。
plt.ylim([-0.5,1.5])
# 軸名をセットします。fontsize=20で字の大きさを20にします。
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
# グラフに説明を加えます。
plt.legend(fontsize=20)
# プロットします。
plt.show()
```

## パーセプトロンの学習


$$
\vec{x}=(x,y),\vec{w}=(w_{1},w_{2})
$$

$$
\phi(x,y)=w_{0}+w_{1}x+w_{2}y
$$

パーセプトロンは次のルールで学習するとします。$\delta$はマージンと呼ばれたりもします。判定での「余裕」です。

- Case I:$$

$$
\phi(x,y)>\delta \,\,\, \mbox{and} \,\,\, \mbox{OUT}=1  \to \mbox{No Change} 
$$

- Case II:
$$
\phi(x,y)<-\delta \,\,\, \mbox{and} \,\,\, \mbox{OUT}=0  \to \mbox{No Change} 
$$

- Case III:

$$
\phi(x,y)<\delta \,\,\, \mbox{and} \,\,\, \mbox{OUT}=1  \to 
\vec{w}=\vec{w}+\vec{x},w_{0}=w_{0}+1 
$$

- Case IV:

$$
\phi(x,y)>-\delta \,\,\, \mbox{and} \,\,\, \mbox{OUT}=0  \to  
\vec{w}=\vec{w}-\vec{x},w_{0}=w_{0}-1
$$


$\vec{x},\vec{w},w_{0}$をインプットして、Case I, Case IIの場合に0、Case IIIの場合に1,Case IVの場合に-1を出力する関数Judge$(\vec{x},\vec{w},w_0$,out,$\delta$)を考えます。

$$
\mbox{Judge}(\vec{x},\vec{w},w_0,\mbox{out})
=\left\{
\begin{array}{cc}
0 &  \vec{w}\cdot \vec{x}+w_{0}>\delta \,\,\mbox{and}\,\, \mbox{out}=1 \\
0 &  \vec{w}\cdot \vec{x}+w_{0}<-\delta \,\,\mbox{and}\,\, \mbox{out}=0 \\
1  &   \vec{w}\cdot \vec{x}+w_{0}<\delta \,\,\mbox{and}\,\, \mbox{out}=1   \\
-1 &  \vec{w}\cdot \vec{x}+w_{0}>-\delta \,\,\mbox{and}\,\, \mbox{out}=-1  \\
\end{array}
\right.  
$$

すると、パーセプトロンの学習は

$$
\vec{w}\to \vec{w}+\mbox{Judge}(\vec{x},\vec{w},w_{0},\mbox{out})\times \vec{x},w_{0}\to w_{0}+\mbox{Judge}
$$

と書くことができます。


```python
def judge(x,w,w_0,out,delta):
  # 出力を0にします。CaseI,IIでこれを出力します。
  result=0
  # Case IIIの場合です。
  if np.dot(x,w)+w_0<delta and out==1:
    result=1
  # Case IVの場合です。
  if np.dot(x,w)+w_0>-delta and out==0:
    result=-1
  return result
```

では、$\vec{w}=(0,0),w_{0}=0$から出発してORデータを正しく判定できる$\vec{w},w_{0}$を獲得できるのか学習してみます。


```python
w=np.array([0,0])
w_0=0
for i in range(1000):
  j=i%100
  result=judge(XY100[j,:],w,w_0,OUT100[j],0.1)
  w=w+result*XY100[j,:]
  w_0=w_0+result
  print(i+1,result,w,w_0)
```

## Irisデータの識別・分類

Irisデータセットは「あやめ」という花の形状データで、「花びら（petal）／がく片（sepal）の長さと幅（cm）」という4つの特徴量と、あやめの種類である「setosa／versicolor／virginica」という3つのラベルで構成されるデータセットである。

- Author： R.A. Fisher

- Title： The use of multiple measurements in taxonomic problems, Annual Eugenics, 7, Part II, 179-188.

- year： 1936

[URL](https://archive-beta.ics.uci.edu/ml/datasets/iris)

iris.png

```python
# サンプルデータの読み込み
from sklearn.datasets import load_iris
# "Sepal_Length","Sepal_Width","Petal_Length","Petal_Width"のうち、Sepal Length, Petal Lengthのみ
# "Setosa","Versicolor","Virginica"のうちsetosa,versicolorのデータのみ
XY_iris = load_iris().data[0:100,[0,2]] 
# "Setosa","Versicolor","Virginica"のうちsetosa,versicolorのデータのみ
OUT_iris= load_iris().target[0:100]
# 最初の50はsetosa=0,次の50はversicolor=1
```

### 課題

XY_irsiに100行2列のデータ,OUT_irisに0，1のラベルがある。これを用いてパーセプトロンを学習し、アヤメを分類してみよう。

```python
# プロットする図の大きさをfigsize=(12,12)で指定します。
plt.figure(figsize=(12,12)) 
# OUT=1に対応するXY100の行のデータを2次元面にプロットします。
# xにXY_iris[,0],yにXY_irsi[,1]をインプット、maker="s"としてマーカーを"四角",サイズをs=100、色をcolor="red"と指定します。
plt.scatter(x=XY_iris[OUT_iris==1,0],y=XY_iris[OUT_iris==1,1],label='OUT=1(Versicolor)',marker="s",s=100,color="red")
# OUT=0に対応するXY100の行のデータを2次元面にプロットします。
# xにXY_irsi[,0],yにXY_irsi[,1]をインプット、maker="o"としてマーカーを"丸",サイズをs=100、色をcolor="blue"と指定します。
plt.scatter(x=XY_iris[OUT_iris==0,0],y=XY_iris[OUT_iris==0,1],label='OUT=0(Setosa)',marker="o",s=100,color="blue")
# 表示するx軸の範囲を指定します。
plt.xlim([3,8])
# 表示するy軸の範囲を指定します。
plt.ylim([0,6])
# 軸名をセットします。fontsize=20で字の大きさを20にします。
plt.xlabel("x",fontsize=20)
plt.ylabel("y",fontsize=20)
# グラフに説明を加えます。
plt.legend(fontsize=20)
# プロットします。
plt.show()
```
